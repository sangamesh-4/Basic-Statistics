 ## Problem Statement : Hospital Patient Data Analysis

## Context:
A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. You have two datasets: one with patient info and another with billing details. Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
patient=pd.read_csv('Patient_Data.csv')
patient
```

```python
patient.info()
```

```python
patient.columns
```

```python
patient_imp=patient[['PatientID', 'Department', 'Doctor', 'BillAmount']]
patient_imp.head()
```

```python
patient_droped=patient.drop(['ReceptionistID', 'CheckInTime'],axis=1)
patient_droped
```

```python
patient['Department'].unique()
```

```python
patient['BillAmount'].unique()
```

```python
department_total=patient.groupby('Department')['BillAmount'].sum()
department_total = department_total.sort_values(ascending=False)
print(department_total)
```

```python
patientId_duplicate=patient[patient.duplicated(subset='PatientID')]
print(patientId_duplicate)


patient_cleaned=patient.drop_duplicates()
patient_cleaned
```

```python
patient['BillAmount'].isnull()
```

```python
patient['BillAmount']=patient['BillAmount'].fillna(patient['BillAmount'].mean())
```

```python
patient.head()
```

```python
billing=pd.read_csv('Billing_Data.csv')
billing.head()
```

```python
merged_data=pd.merge(patient_cleaned,billing,on='PatientID',how='inner')
merged_data
```

```python
new_patients = pd.DataFrame({
    'PatientID':[106, 107],
    'Department':['Opthalmology', 'Urology'],
    'Doctor':['Dr. Neela', 'Dr. Naveen'],
    'BillAmount':[6500, 7200]
})
```

```python
patient_up=pd.concat([patient_cleaned,new_patients],ignore_index=True)
patient_up
```

```python
patient_up.head()
```

```python
new_billing_cols = pd.DataFrame({
    'InsuranceCovered': [True, False, True, True, False],
    'FinalAmount': [4500, 7200, 7000, 6000, 5800]
})
```
