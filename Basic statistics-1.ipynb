### Descriptive Analytics for Numerical Columns 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('sales_data_with_discounts.csv')
data.head()
```

```python
data.info()
```

```python
num_data=data._get_numeric_data()
num_data.head()
```

```python
mean_data=num_data.mean()
mean_data
```

```python
median_data=num_data.median()
median_data
```

```python
mode_data=num_data.mode()
mode_data.head()
```

```python
std_data=num_data.std()
std_data
```

```python
data.describe()
```

### Data Visualization ##

```python
for col in num_data:
    plt.figure(figsize=(3,2))
    sns.histplot(data[col],kde=True,bins=20)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.boxplot(data["Volume"])
plt.title("Boxplot of Volume")
plt.ylabel("Volume")
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.boxplot(data["Avg Price"])
plt.title("Boxplot of Avg Price")
plt.ylabel("Avg Price")
plt.show()
```

-->In the case of Average Price, the spread of values is quite large. A small number of very high prices are far away from the main concentration of data. These extreme values increase the overall variation and indicate that some products are priced significantly higher than others.

```python
plt.figure(figsize=(4,3))
plt.boxplot(data["Total Sales Value"])
plt.title("Boxplot of Total Sales Value")
plt.ylabel("Total Sales Value")
plt.show()
```

-->The Total Sales Value also shows a wide range of values with several extreme observations. The majority of sales transactions fall within a lower range, but a few very high transactions increase the upper spread of the distribution. This indicates that total revenue is influenced by a limited number of large sales.

```python
plt.figure(figsize=(4,3))
plt.boxplot(data["Discount Rate (%)"])
plt.title("Boxplot of Discount Rate (%)")
plt.ylabel("Discount Rate (%)")
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.boxplot(data["Discount Amount"])
plt.title("Boxplot of Discount Amount")
plt.ylabel("Discount Amount")
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.boxplot(data["Net Sales Value"])
plt.title("Boxplot of Net Sales Value")
plt.ylabel("Net Sales Value")
plt.show()
```

```python
data.boxplot(figsize=(12,5))
plt.title("Boxplot of Numerical Variables")
plt.xlabel("Variables")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.show()
```

```python
categorical_cols = data.select_dtypes(include=['object'])
categorical_cols
```

```python
categorical_cols.head()
```

```python
date_counts = data["Date"].value_counts()
plt.figure(figsize=(8,4))
plt.bar(date_counts.index, date_counts.values)
plt.title("Bar Chart of Date")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
day_counts = data["Day"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(day_counts.index, day_counts.values)
plt.title("Bar Chart of Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
sku_counts = data["SKU"].value_counts()
plt.figure(figsize=(10,4))
plt.bar(sku_counts.index, sku_counts.values)
plt.title("Bar Chart of SKU")
plt.xlabel("SKU")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
bu_counts = data["BU"].value_counts()
plt.figure(figsize=(4,2))
plt.bar(bu_counts.index, bu_counts.values)
plt.title("Bar Chart of BU")
plt.xlabel("BU")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
brand_counts = data["Brand"].value_counts()
plt.figure(figsize=(5,3))
plt.bar(brand_counts.index, brand_counts.values)
plt.title("Bar Chart of Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
model_counts = data["Model"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(model_counts.index, model_counts.values)
plt.title("Bar Chart of Model")
plt.xlabel("Model")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
for col in categorical_cols.columns:
    value_counts = data[col].value_counts()
    plt.figure(figsize=(8,4))
    plt.bar(value_counts.index, value_counts.values)
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
